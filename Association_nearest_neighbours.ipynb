```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
import qgrid
```

```python
data = pd.read_excel("order_products__prior.xlsx",index_col=0,sheetname=1)
data.head()
```

```python
frequent_itemsets = apriori(data, min_support=0.007, use_colnames=True)
```

```python
type(frequent_itemsets)
```

```python
frequent_itemsets.head()
```

```python
frequent_itemsets.info()
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
rules.head()
```

```python
rules.info()
```

```python
rules[rules["lift"]>1]
```

```python
qgrid.show_grid(rules)
```

```python
rules.to_excel("Gariz_lift.xlsx")
```
